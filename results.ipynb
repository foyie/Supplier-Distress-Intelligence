```python
import pandas as pd
import matplotlib.pyplot as plt

forecasts = pd.read_parquet('data/forecasts/all_forecasts.parquet')
signals   = pd.read_parquet('data/processed/feature_matrix_full.parquet')

# Get historical + forecast for company 1's sentiment
hist = signals[signals['company_id'] == 1].sort_values('date')[['date', 'news_sentiment_score']]
fc   = forecasts[(forecasts['company_id'] == 1) & (forecasts['feature'] == 'news_sentiment_score_forecast')]

print("Historical signal:")
print(hist.tail(5))
print("\nForecast:")
print(fc[['date', 'value', 'value_lower', 'value_upper']])
```
